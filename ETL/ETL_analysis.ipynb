### Transform raw data to processed for analysis

```python
dataset_analysis = df.copy()
remove_cols = ['country', 'region', 'specificity', 'vicinity', 'summary', 'attacktype1', 'targtype1', 'targsubtype1', 'natlty1', 'nperps', 'nperpcap', 'claimed', 'weaptype1', 'weaptype1_txt',
               'weapsubtype1', 'weapsubtype1_txt', 'weapdetail', 'property', 'scite1', 'dbsource', 'INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY']
dataset_analysis = dataset_analysis.drop(columns=remove_cols)
dataset_analysis.head()
```

```python
df['date'] = pd.to_datetime({"year": df['iyear'], "month": df['imonth'], "day": df['iday']}, errors="coerce")
df.info()
```

### Separate the dataframe into decades

```python
def df_decades(df, year: int):
    # return a dataframe with all the terrorism data for each decade starting from 1970 to 2017
    if year in df.iyear:
        dataF = pd.DataFrame()
        for i in range(year, year+9):
            data = pd.DataFrame
            data = df[df['iyear'] == i]
            dataF = pd.concat([dataF, data])
        return dataF
    else:
        print('Year Not found')
    # Only add the beggin of the decade 1970, 1980, 1990...
```

```python
for i in range(1970, 2017, 10):
    df_name = f"df_{i}"
    df_temp = df_decades(dataset_analysis, i)
    df_temp.to_csv(f'../datasets/datasets_processed/datasets_for_analysis/{df_name}.csv')
```
